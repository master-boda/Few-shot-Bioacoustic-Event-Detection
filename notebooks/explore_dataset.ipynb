# Dataset exploration

Quick inspection of class counts and event durations for DCASE Task 5 data.

```python
from pathlib import Path
import pandas as pd
from collections import Counter

data_root = Path("../data/raw/development/Development_Set")
train_meta = data_root / "Training_Set"
val_meta = data_root / "Validation_Set"

def load_annotations(root: Path):
    rows = []
    for csv_path in root.rglob("*.csv"):
        df = pd.read_csv(csv_path)
        rows.append((csv_path.name, len(df)))
    return rows

train_rows = load_annotations(train_meta)
val_rows = load_annotations(val_meta)
print("Training files:", len(train_rows))
print("Validation files:", len(val_rows))
print("Train events per file (min/mean/max):", min(r[1] for r in train_rows),
      sum(r[1] for r in train_rows)/len(train_rows), max(r[1] for r in train_rows))
print("Val events per file (min/mean/max):", min(r[1] for r in val_rows),
      sum(r[1] for r in val_rows)/len(val_rows), max(r[1] for r in val_rows))
```
